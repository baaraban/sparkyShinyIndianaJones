```python
dates = []

for index, row in result_df.iterrows():
    x = row['transitions']
    x = x.replace(', {', '{')
    to_delete = ['[', ']', '{}']
    for ch in to_delete:
        x = x.replace(ch, '')
    x = x.replace('}{', '}\t{')
    x = x.split('\t')
    dates.extend([elem.split(':')[0][1:].strip() for elem in x if elem])
```

```python
def try_date_parsing_method(dates, parse_func):
    parsed = {}
    unparsed = []
    
    for x in dates:
        try:
            p = parse_func(x)
            parsed[x] = p
        except:
            unparsed.append(x)
    
    print(f"Totally parsed - {len(parsed)}")
    print(f"Totally unparsed - {len(unparsed)}")
    print("_______")
    print("Parsed: ")
    for key, value in parsed.items():
        print(f"{key} - {value}")
    print("______")
    print("Not parsed:")
    for x in unparsed:
        print(x)
    
    return unparsed
```
