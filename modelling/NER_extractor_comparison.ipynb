```python
import spacy
from spacy import displacy
import pandas as pd
```

```python
import en_core_web_sm
nlp_sm = en_core_web_sm.load()

import en_core_web_md
nlp_md = en_core_web_md.load()

import en_core_web_lg
nlp_lg = en_core_web_lg.load()
```

```python
PATH_TO_CSV = "../data_collection/museums_scraping/raw_data/louvre_artifacts.csv"
```

```python
df = pd.read_csv(PATH_TO_CSV).drop(['Unnamed: 0'], axis=1)
```

```python
df.head()
```

```python
for index, row in df.iterrows():
    working_str = row['acquired_by'].strip()
    print(f"{index} - {working_str}")
    
    small = [(X.text, X.label_) for X in nlp_sm(working_str).ents]
    medium = [(X.text, X.label_) for X in nlp_md(working_str).ents]
    large = [(X.text, X.label_) for X in nlp_lg(working_str).ents]
    
    print("_______________")
    print(f"Small - {small}")
    print("_______________")
    print(f"Medium - {medium}")
    print("_______________")
    print(f"Large - {large}")
    print("_______________\n")
```
