```python
from polyglot.text import Text
import pandas as pd

PATH_TO_CSV = "../data_collection/museums_scraping/raw_data/louvre_artifacts.csv"

df = pd.read_csv(PATH_TO_CSV).drop(['Unnamed: 0'], axis=1)

df.head()
```

```python
def run_NER_per_column(df, column):
    for index, row in df.iterrows():
        working_str = row[column].strip()
        print(f"{index} - {working_str}")
        
        text = Text(working_str)
        text.language = 'en'
        entities = [(' '.join(X), X.tag) for X in text.entities]
    
        print("_______________")
        print(f"Ents - {entities}")
        print("_______________\n")
```

```python
run_NER_per_column(df, 'acquired_by')
```
